**Normalization/Scaling:** <br>
 - MinMaxScaler, StandardScaler, etc.: Scale numerical features to a specific range or with a standard deviation.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(data[['ApplicantIncome', 'CoapplicantIncome']])
data[['ApplicantIncome', 'CoapplicantIncome']] = data_scaled
data
```

# 4. Feature Engineering
