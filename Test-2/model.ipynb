```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.neighbors import NearestNeighbors
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples, davies_bouldin_score, calinski_harabasz_score
```

# Feature Engineering

```python
features = df[['child_mort', 'exports', 'health', 'imports', 'income', 'inflation', 'life_expec', 'total_fer', 'gdpp']]
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

# K-Means Clustering

```python
kmeans = KMeans(n_clusters=3, random_state=0)
df['cluster'] = kmeans.fit_predict(scaled_features)
centers = scaler.inverse_transform(kmeans.cluster_centers_)
center_df = pd.DataFrame(centers, columns=features.columns)
center_df['cluster'] = range(len(centers))

```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='income', y='life_expec', hue='cluster', palette='viridis', s=100)
sns.scatterplot(data=center_df, x='income', y='life_expec', hue='cluster', palette='dark', s=300, markers='X')
plt.title('K-means Clustering of Countries')
plt.xlabel('Income')
plt.ylabel('Life Expectancy')
plt.legend(title='Cluster')
plt.show()
```

## Gradient Boosting Classifier

```python

from sklearn.ensemble import GradientBoostingClassifier
import seaborn as sns

df = pd.read_csv('Country-data.xls')
df.head()
```

```python
median_income = df['income'].median()
df['label'] = (df['income'] > median_income).astype(int)

features = df[['child_mort', 'exports', 'health', 'imports', 'income', 'inflation', 'life_expec', 'total_fer', 'gdpp']]
labels = df['label']
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
gb_clf = GradientBoostingClassifier(n_estimators=100, random_state=0)
gb_clf.fit(scaled_features, labels)

df['predicted_label'] = gb_clf.predict(scaled_features)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='income', y='life_expec', hue='predicted_label', palette='viridis', s=100)
plt.title('Gradient Boosting Classification of Countries')
plt.xlabel('Income')
plt.ylabel('Life Expectancy')
plt.legend(title='Predicted Label')
plt.show()
```
